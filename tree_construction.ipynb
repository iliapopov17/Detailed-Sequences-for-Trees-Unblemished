## **Step 1. Combine all the sequences to one file**

```python
! cat genbank_sequences/*.fasta > demo_data/all_seqs.fa
```

### **Check the number of sequences**

```python
with open("demo_data/all_seqs.fa", "r") as fasta_file:
    content = fasta_file.read()
    num_sequences = content.count(">")
print(f"The number of sequences in combined file: {num_sequences}")
```

## **Step 2. Multiple Sequence Alignment**

```python
! mafft --auto demo_data/all_seqs.fa > demo_data/all_seqs_mafft.fa
```

## **Step 3. Launching ModelFinder to find the best substitution model**

```python
! iqtree2 -m MFP -s demo_data/all_seqs_mafft.fa --prefix demo_data/tree_MF2 -T AUTO
```

### **Getting the best substitution model**

```python
! head -42 demo_data/tree_MF2.iqtree | tail -6
```

## **Step 4. Launching IQ-TREE to construct the tree**

```python
! iqtree2 -s demo_data/all_seqs_mafft.fa -m GTR+F+R7 -pre demo_data/tree_ufb -bb 1000 -nt AUTO
```
